```python
import requests
import re
import json
```

```python
research = set()
course = set()
edu_project = set()
examples = set()
service = set()
phd = set()
master = set()
bachelor = set()
```

```python
with open('../profile/reps_courses.md') as inp:
    text = inp.read()
    reps = re.findall(re.escape('https://github.com/intelligent-Systems-Phystech/')+'[^\s\)\]]+', text, re.IGNORECASE)
for r in reps:
    course.add(r)
```

```python
with open('../profile/reps_research.md') as inp:
    text = inp.read()
    reps = re.findall(re.escape('https://github.com/intelligent-Systems-Phystech/')+'[^\s\)\]]+', text, re.IGNORECASE)
for r in reps:
    research.add(r)
```

```python
with open('../profile/reps_edu.md') as inp:
    text = inp.read()
    reps = re.findall(re.escape('https://github.com/intelligent-Systems-Phystech/')+'[^\s\)\]]+', text, re.IGNORECASE)
for r in reps:
    edu_project.add(r)
```

```python
with open('../profile/reps_other.md') as inp:
    text = re.split('##', inp.read())[1]
    reps = re.findall(re.escape('https://github.com/intelligent-Systems-Phystech/')+'[^\s\)\]]+', text, re.IGNORECASE)
for r in reps:
    examples.add(r)
```

```python
with open('../profile/reps_other.md') as inp:
    text = re.split('##', inp.read())[2]
    reps = re.findall(re.escape('https://github.com/intelligent-Systems-Phystech/')+'[^\s\)\]]+', text, re.IGNORECASE)
for r in reps:
    service.add(r)
```

```python
with open('../profile/reps_students.md') as inp:
    for line in inp:
        found = re.search(re.escape('https://github.com/intelligent-Systems-Phystech/')+'[^\s\)\]]+', line, re.IGNORECASE)
        if found is not None:
            if 'bachelor' in line.lower():
                bachelor.add(found.group())
            elif 'phd' in line.lower():
                phd.add(found.group())
            else:
                master.add(found.group())

            
```

```python
USER=''
API_TOKEN=''
GIT_API_URL='https://api.github.com'
```

```python
def update_topics(repo_url, new_topics):
    rep_name = repo_url.strip('/').split('/')[-1]
    url = '/repos/Intelligent-Systems-Phystech/{}/topics'.format(rep_name)
    print (GIT_API_URL+url)
    request = requests.get(GIT_API_URL + url, auth=(USER, API_TOKEN))  
    json = request.json()    
    topics = json.get('names', [])
    
    new_topics = topics + new_topics
    json = {
        "names":new_topics,
    }
    request = requests.put(GIT_API_URL + url, auth=(USER, API_TOKEN), json = json)  
    
 
```

```python
for r in course:
    update_topics(r, ['course'])
```

```python
for r in research:
    update_topics(r, ['research'])
```

```python
for r in edu_project:
    update_topics(r, ['edu-project'])
```

```python
for r in service:
    update_topics(r, ['service'])
```

```python
for r in phd:
    update_topics(r, ['graduation', 'phd'])
```

```python
for r in bachelor:
    update_topics(r, ['graduation', 'bachelor'])
```

```python
for r in master:
    update_topics(r, ['graduation', 'master'])
```
