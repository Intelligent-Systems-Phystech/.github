```python
import requests
import re
import json
ru = re.compile('[А-Яа-яЁё]')
```

```python
# repositories were manually gathered. 
# In principle they can be gathered by api
with open('reps.txt') as inp:
    reps = inp.readlines()
reps = [r.strip() for r in reps]
```

```python
results = {}
```

```python
# gathering 
```

```python
USER=''
API_TOKEN=''
GIT_API_URL='https://api.github.com'
```

```python
# description
def get_descr(rep_name):
    try:
        url = '/repos/Intelligent-Systems-Phystech/{}'.format(rep_name)
        print (GIT_API_URL+url)
        request = requests.get(GIT_API_URL + url)  
        
        json = request.json()
        return json
    except Exception as e:
        print ('Failed to get api request from %s' % url, repr(e))
for r in reps:
    if r in results:
        continue
        
    d = get_descr(r)
    if d is not None and d.get('description') is not None:
        if ru.match(d['description']):
            results[r] = {'ru-descr':d['description']}
        else:
            results[r] = {'en-descr':d['description']}
```

```python
# contributors

def get_contr(rep_name):
    try:
        url = '/repos/Intelligent-Systems-Phystech/{}/contributors'.format(rep_name)
        print (GIT_API_URL+url)
        request = requests.get(GIT_API_URL + url, auth=(USER, API_TOKEN))
        json = request.json()
        return json
    except Exception as e:
        print ('Failed to get api request from %s' % url, repr(e))
for r in reps:    
    d = get_contr(r)
    if d is not None and len(d)>0:
        if r not in results:
            results[r] = {}
        results[r]['users'] = [u['login'] for u in d]
            
        
```

```python
with open('rep_results.json', 'w') as out:
    out.write(json.dumps(results))
```

```python
all_users_nicks = set()
for r in results:
    all_users_nicks = all_users_nicks.union(results[r]['users'])
```

```python
simple_cases = []
for n in sorted(all_users_nicks):
    if n.count('-'):
        simple_cases.append(n.replace('-', ' '))
        continue
    upper = [i for i in range(len(n)) if n[i].isupper()]
    if len(upper) == 2 and upper[0] == 0:
        simple_cases.append(n[:upper[1]]+' '+n[upper[1]:])
        continue
    simple_cases.append('???')
```

```python
print ('\n'.join(simple_cases))
```

```python
gh_nicks = {}
with open('gh_nicks.csv') as inp:
    for line in inp:
        tokens = line.strip().split(',')
        gh_nicks[tokens[0]]= tokens[1]
```

```python
groups = sorted(set([r.replace('_', '-').split('-')[0] for r in reps]))[::-1]
```

```python
for g in groups:
    subitems = sorted([r for r in reps if r.replace('_', '-').split('-')[0] == g])
    print ('## '+g)
    print ('Repository description|Repository description (Russian)|Link| Contributors |')    
    print ('|--- | --- | --- | --- |')
    for r in subitems:
        contributor_buf = []
        for contr in results[r]['users']:
            if len(contributor_buf)>0:
                contributor_buf.append(' , ')
            try:
                contributor_buf.append('[{}](https://github.com/{})'.format(gh_nicks[contr], contr))
            except KeyError:
                contributor_buf.append('[{} (???)](https://github.com/{})'.format(contr, contr))
            
        print ('| {} | {} | [{}](https://github.com/intelligent-Systems-Phystech/{}) | {} |'.format(results[r].get('en-descr','-'),
                                                   results[r].get('ru-descr','-'),
                                                   r, r,
                                                   ''.join(contributor_buf)))
    
```
